```python
import seaborn as sns
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
data = pd.DataFrame(data = [[10,0],[12,0],[10.8,0],[11.2,0],[18,1],[17,1],[13.4,1],[16,0],[14,0],[20,1],[11,0],[12,0],[17,1],[17,1],[14,1],[19.3,1],[19,1],[20,1],[18,1]], columns = ['Variaveis que indicam facilidade do exercicio', 'Probabilidade Acerto'])
```

```python
ax = sns.regplot(y = 'Probabilidade Acerto', x = 'Variaveis que indicam facilidade do exercicio', data=data, logistic=True)
```

```python
xx = ax.get_lines()[0].get_xdata()
yy = ax.get_lines()[0].get_ydata()
```

```python
fig, ax = plt.subplots(figsize = (4, 4), dpi = 150)
ax.plot(xx,yy)
ax.scatter(xx[37],yy[37], c='r', label = str(round(100*yy[37],2)) + ' %')
ax.scatter(xx[75],yy[75], c='g', label = str(round(100*yy[75],2)) + ' %')
ax.scatter(xx[93],yy[93], c='y', label = str(round(100*yy[93],2)) + ' %')
ax.set_xlabel('Variaveis que indicam facilidade do exercicio')
ax.set_ylabel('Probabilidade Acerto')
fig.suptitle('Dados de teste na Regressao Logistica calculada')
ax.legend()
```
