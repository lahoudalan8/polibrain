```python
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
import math
from numpy import inf
```

```python
ca1 = 0.8*np.ones(300)
ca2 = 0.5*np.ones(300)
ca3 = 0.5*np.ones(300)

ct1 = 100*np.ones(300)
ct2 = 100*np.ones(300)
ct3 = 200*np.ones(300)
```

```python
t = np.arange(1,301,1)
```

```python
f1 = np.divide((1-ca1),((np.log10(np.minimum(ct1,t)))))
f2 = np.divide((1-ca2),((np.log10(np.minimum(ct2,t)))))
f3 = np.divide((1-ca3),((np.log10(np.minimum(ct3,t)))))
```

```python
fig, ax = plt.subplots(figsize = (4,4), dpi = 150)
ax.plot(t, f1, label = 'ca = 0.8; ct = 100s')
ax.plot(t, f2, label = 'ca = 0.5; ct = 100s')
ax.plot(t, f3, label = 'ca = 0.5; ct = 200s')
ax.set_xlabel('Tempo no exercicio(s)')
ax.set_ylabel('Pontuacao no exercicio')
ax.legend()
fig.suptitle('Pontuacao por exercicio acertado')
```
